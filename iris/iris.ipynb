```python
#2
import pandas as pd
pd.set_option("display.max_columns", 100)
url = "https://raw.githubusercontent.com/cs6220/cs6220.spring2019/master/data/AmesHousing.txt"
df2 = pd.read_csv(url, sep="\t")
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
df_missing = df2.isna().sum()
```

```python
df_missing
```

```python
count = 0
for i in df_missing:
    if(i > 0):
        count += 1
count
#2.1.1 there are 27 features have missing values.
```

```python
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
```

```python
df2
for each_column in df2:
    if(is_string_dtype(df2[each_column])):
        df2[each_column] = df2[each_column].fillna("Missing")
        continue
df2
#2.1.2
```

```python
for each_column in df2:
    if(is_numeric_dtype(df2[each_column])):
        df2[each_column] = df2[each_column].interpolate(method ='linear')
        continue
df2

#2.1.3

```

```python
df2.boxplot(figsize=(40,40))
```

```python
#2.2.1
from sklearn import preprocessing
for each_column in df2:
    if(is_numeric_dtype(df2[each_column])):
        df2[each_column] = preprocessing.scale(df2[each_column])
        continue
df2
```

```python
#2.2.2
df2.boxplot(figsize=(40,40))
```

```python
#As can be seen the circle is way more spread out than the plot before standardization. The size of the box
#is also different for each feature after standardization. That means the data is standardized so there 
#will be unit variance and zero mean. As we can see, feature order has the outlier furtherst from mean before
#standardization, misc val has the outlier furtherst from the mean after standardization.
```

```python
#2.3
from sklearn import feature_selection
```

```python
import numpy as np
lst = []
for each_column in df2:
    if(is_numeric_dtype(df2[each_column])):
        info = feature_selection.mutual_info_regression(np.array(df2['SalePrice']).reshape(-1, 1), np.array(df2[each_column]))
        lst.append((each_column,info[0]))
        continue
lst
#2.3.1
```

```python
lst.sort(key = lambda x: x[1],reverse=True)
```

```python
#2.3.2
lst
```

```python
#As we can see, except for salePrice itself, the top 5 numeric features ranked by mutual information are
#Overall Qual, Gr Liv Area, Garage Area, Year Built, Total Bsmt SF
```

```python
#2.3.3
#I would expect the price goes up if these top ranked values go up and vice versa. Because according to
#the mutual information formula, if the MI goes up, joint probability will also go up.
```

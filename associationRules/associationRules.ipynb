```python
#1.1
import pandas as pd
pd.set_option("max_colwidth", 150)
f = "https://github.com/cs6220/cs6220.spring2019/raw/master/data/Online%20Retail.xlsx"
df = pd.read_excel(f)
basket = (df[df["Country"] == "United Kingdom"]
.groupby(["InvoiceNo", "Description"])["Quantity"]
.sum().unstack().reset_index().fillna(0)
.set_index("InvoiceNo")) # transform transactions into baskets of items
basket_sets = basket.applymap(lambda x: 1 if x >=1 else 0) # convert counts to booleans
df
```

```python
basket_sets
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
basket_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True)
basket_itemsets
```

```python
oneItemSets = basket_itemsets[basket_itemsets.itemsets.map(len)==1]
twoItemSets = basket_itemsets[basket_itemsets.itemsets.map(len)==2]
#basket_itemsets.sort_values(by=['support'], ascending=False)
#for i in basket_itemsets["itemsets"]:
    #print(len(i))
```

```python
#2
import numpy as np
import pandas as pd
path = "https://raw.githubusercontent.com/cs6220/cs6220.spring2019/master/data/adult/"
names = pd.read_csv(path + "adult.names", sep="\n", header=None)
parse_cols = lambda x: x.str.split(":", expand=True).iloc[:, 0]
columns = np.roll(parse_cols(names.iloc[92:108, 0]), shift=-1)
df_adult = pd.read_csv(path + "adult.data", sep=",", header=None, index_col=False)
df_adult.columns = columns
```

```python
df_adult
```

```python
df_nominal = df_adult.select_dtypes(exclude=[np.number])
df_nominal
```

```python
#2.1.1
oneHot_encoded = pd.get_dummies(df_nominal)
oneHot_encoded
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
itemsets = apriori(oneHot_encoded, min_support=0.05, use_colnames=True, max_len=3)
itemsets
```

```python
#2.1.2
#I used min support of 0.05 so I can have fair amount(504 rows) of data instead of too little or too big.
#Then I used confidence with min threshold of 0.1 even thought I sort the df by confidence.
#Here are rules I found interesting:
#1. row 1809 (relationship_ Husband, >50K, <=50K._ >50K) (sex_ Male), this rule has the highest confidence 
#of 1. And it makes sense since it's a husband with more than 50k annual income, then the consequents of being
#a male is very likely.
#2. Rows like 295, with missing information with every part, it still has the highest confidence yet the 
#antecedent support and consequent support are low.
#3. Row 1098 (relationship_ Husband, education_ Some-college), (marital-status_ Married-civ-spouse) this rule
#also makes sense, since it's a husband that went to college, the marital status is married.
#4. Row 1042 (education_ HS-grad, relationship_ Not-in-family) and (>50K, <=50K._ <=50K), in general HS-grad
#with less than 50k annual income makes sense.
#5. For the first 60 rows, which are the top 60 in confidence. Rules assoicated with annual income are usually
#less than 50k. I do not really know the reasoning behind this.
ar21 = association_rules(itemsets, metric="confidence", min_threshold=0.1)
ar21.sort_values('confidence', ascending=False).head(60)
```

```python
#2.1.3
#1. For row 1618, (marital-status_ Never-married, sex_ Male) (relationship_ Own-child). It's a small chance
#but it has high lift. And there are also rules with same parts but on the opposite position. I assume this is 
#due to the min support value choose is  too low. So rules do not really make sense also appear.
#2. For row 983, It displays a rule between education, relationship and income. Bachelor degree is assoicated
#with higher annual income makes sense. But relationship being husband does not seem fit in this rule. It's
#probably still beacuse min support is too low.
#3.For row 2151, there is a rule between (relationship_ Unmarried) and (>50K, <=50K._ <=50K, sex_ Female),
#rules like this also appeared in top rules with meteric confidence. In top results, there are many rules 
#between gender and annual income which are usually female with less than 50k annual income which may reflect
#some social issues. 
#4. For row 1838, there is a rule between 	(>50K, <=50K._ >50K) and (sex_ Male, occupation_ Exec-managerial)
#which describes mroe than 50k annual income and gender is male with a exec-managerial job, which I believe it 
#reflect the relation bewteen income and gender and occupation. Combining with information from above, it should
#reflect some social issues if the data is real.
#5. The top rules by using lift metric are usually between gender and some other features, and some rules 
#does not fall under common sense. I believe it's because of min support is too low.
ar22 = association_rules(itemsets, metric="lift", min_threshold=0.1)
ar22.sort_values('lift', ascending=False).head(60)
```

```python
#2.1.4
f_itemsets1 = apriori(oneHot_encoded, min_support=0.02, use_colnames=True, max_len=3)
f_itemsets1
i1 = association_rules(f_itemsets1, metric="confidence", min_threshold=0.1)
i2 = association_rules(f_itemsets1, metric="lift", min_threshold=0.1)
i1.sort_values('confidence', ascending=False)
```
